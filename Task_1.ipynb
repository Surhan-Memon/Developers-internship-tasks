```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#import data
iris = pd.read_csv('iris.csv')
```

```python
#display dataset structure using .shape:
iris.shape
```

```python
#display dataset structure using .columns:
iris.columns
```

```python
#display dataset structure using .head():
iris.head()
```

```python
#scatter plot to analyze the relationship between variables
sns.scatterplot(x =iris['sepal_length'], y=iris['sepal_width'], hue=iris['species'])
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.title('scatter plot between sepal_length and sepal_width')
plt.show()
```

```python
#histogram to examine data distribution
sns.histplot(iris['sepal_length'], bins= 30, alpha = 0.4, color = 'yellow' )
plt.xlabel('sepal_length')
plt.ylabel('frequency')
plt.title('histogram of sepal_length')
plt.show()
```

```python
#box plot to detect outliers and spread of values
sns.boxplot(x = iris['species'], y = iris['sepal_length'])
plt.xlabel('species')
plt.ylabel('sepal_length')
plt.title('box plot of sepal_length')
plt.show()
```
